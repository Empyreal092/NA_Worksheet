```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x : np.exp(np.sin(x)+np.exp(1)*np.cos(x))+np.exp(x)
```

```python
x_ary = np.linspace(0,2*np.pi,1000000)

x_ary = x_ary
h = x_ary[1]-x_ary[0]

f_ary = f(x_ary)

truth = np.sum(f_ary[1:-1])*h
truth += (f_ary[0]+f_ary[-1])*h/2
```

```python
n_ary = np.arange(3,31)
err_ary = np.empty(np.size(n_ary))

for n in n_ary:
    x_ary = np.linspace(0,2*np.pi,n)
    # print(x_ary)
    
    x_ary = x_ary
    h = 2*np.pi/(n-1)
    
    f_ary = f(x_ary)
    
    trap = np.sum(f_ary[1:-1])*h
    trap += (f_ary[0]+f_ary[-1])*h/2
    
    err = np.abs(trap-truth)
    print(err)
    
    err_ary[n-3] = err
```

```python
plt.loglog(n_ary,err_ary,'o')
plt.loglog(n_ary,1/n_ary**(2)*1e3)
plt.xlabel('$n$')
plt.ylabel('Error')
```
