# Plotting with matplotlib
* [matplotlib intro tutorial](https://matplotlib.org/stable/tutorials/introductory/index.html)
* [matplotlib documentation](https://matplotlib.org/stable/index.html)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,1000)
```

```python
fig = plt.figure(figsize=(4,3))
plt.plot(x,np.sin(x))

plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
plt.plot(x,np.sin(x),label="$\sin(x)$")
plt.plot(x,np.cos(x),label="$\cos(x)$")

plt.ylim(-2, 3)
plt.legend()
plt.title("Some functions")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))

x = np.linspace(0,10,10)
plt.semilogy(
    x, np.exp(x), color='b', linestyle=':', linewidth=2,
    marker='.', markerfacecolor='magenta', markeredgecolor='magenta',
    alpha=0.7
)

plt.title('Semilogy with some some extra styles')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
fig = plt.figure(figsize=(4,3))

x = np.linspace(1, 100)
y1 = x**2
y2 = x
plt.loglog(x, y1, label='$x^{2}$')
plt.loglog(x, y2, label='$x^{1}$')
plt.axvline(10,color='k')

plt.title('loglog plot')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.show()
```

```python
x = np.linspace(0,10,1000)

fig = plt.figure(figsize=(4,3))
plt.plot(x,np.sin(x),label="$\sin(x)$")
plt.plot(x,np.cos(x),label="$\cos(x)$")

plt.ylim(-2, 3)
plt.legend()
plt.title("Some functions")
plt.xlabel("$x$")
plt.ylabel("$y$")

plt.savefig("Example_save.pdf", bbox_inches='tight') # always save lineplots as vector format (.pdf, .eps, etc.)
plt.show()
```
