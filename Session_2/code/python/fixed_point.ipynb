```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2,3,0.01)
```

```python
# print(x)
```

```python
g_1 = lambda x: x**2-2
g_2 = lambda x: -np.sqrt(x+2)
```

```python
plt.figure(figsize=(3,4))

plt.plot(x,g_1(x),label="$g_1(x)$")
plt.plot(x,g_2(x),label="$g_2(x)$")
plt.plot(x,x,'k',label="$y=x$")

plt.legend()
plt.axis('equal')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.ylim([-2.5,4])
```

```python
r = 5
r_save = np.array([5])

for i in range(10):
    print(r)
    r = g_2(r)
    r_save = np.append(r_save,r)
```

```python
plt.plot(np.arange(1,r_save.size+1),r_save,'o')

plt.xlabel("Steps")
plt.ylabel("Roots")
```

```python
plt.semilogy(np.arange(1,r_save.size+1),abs(r_save-2),'o',label="FP Error")
plt.semilogy(np.arange(1,r_save.size+1),(1/4)**np.arange(1,r_save.size+1),'k',label="$(1/4)^k$")

plt.legend()
plt.xlabel("Steps")
plt.ylabel("Error")
```
