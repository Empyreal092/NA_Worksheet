```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
A = np.random.randn(3,3)

print(A)
```

```python
n_ary = 2**np.arange(9)*100
t_ary = np.empty(n_ary.size)

for i in np.arange(n_ary.size):
    
    n = n_ary[i]
    A = np.random.randn(n,n)
    
    t0 = time.time()
    
    l1_norm = np.max(np.sum(np.abs(A),0))
    
    t1 = time.time()
    
    t_ary[i] = t1-t0
    print(t1-t0)
```

```python
plt.loglog(n_ary,t_ary,'o')
plt.loglog(n_ary,n_ary**2/1e8,'k')

plt.xlabel("$n$")
plt.ylabel("Time")
```

```python
n_ary = 2**np.arange(9)*100
t_ary = np.empty(n_ary.size)

for i in np.arange(n_ary.size):
    
    n = n_ary[i]
    A = np.random.randn(n,n)
    
    t0 = time.time()
    
    l1_norm = np.linalg.norm(A, ord=1)
    
    t1 = time.time()
    
    t_ary[i] = t1-t0
    print(t1-t0)
```

```python
plt.loglog(n_ary,t_ary,'o')
plt.loglog(n_ary,n_ary**2/1e8,'k')

plt.xlabel("$n$")
plt.ylabel("Time")
```
